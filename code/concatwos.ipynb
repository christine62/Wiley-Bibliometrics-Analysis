```python
#load txt from folder
from glob2 import glob
file_dir = '../data/Archive/*'
```

```python
files = glob(file_dir)
files
```

```python
#concat dataframe
li = []

for filename in files:
    df = pd.read_csv(filename,sep='\t', encoding='utf-16', error_bad_lines=False, index_col=False)
    df =df[['PT','PY','AF','TI','SO','LA','DT','DE','ID','AB','FU','CR','NR','TC','Z9','U1','U2','PU','PI','PG','WC','SC','UT']]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
