**Load the impact factor file**

```python
dfs = {i: pd.read_csv('../data/impactfactor/JournalHomeGrid{}.csv'.format(i)) for i in range(2010,2018)}
```

```python
def ifpreprocess(df):
    df['Full Journal Title'] = df['Full Journal Title'].str.upper()
    df.drop_duplicates(subset=['Full Journal Title'], keep='first', inplace=True)  
    
    return df.set_index('Full Journal Title')['Journal Impact Factor']
```

```python
for i in dfs:
    dfs[i]=ifpreprocess(dfs[i])
```

```python
conditions=[]
for n in range(2010,2018):
    conditions.append((df['PY']==n))
```

```python
choices = []
for n in range(2010,2018):
    choices.append((df['SO'].map(dfs[n])))
```

```python
df['IM'] = np.select(conditions,choices)
```
