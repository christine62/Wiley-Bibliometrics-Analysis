Build author dataset from 2010 to 2010, authors will be considered by all authors and lead authors

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/wsdata1017withim.csv')
```

```python
df.head()
```

```python
#add funding label
df['fund_label'] = df['FU'].fillna(0).apply(lambda x: 1 if x!=0 else 0)
```

```python
# get author list
df1012 = df[(df['PY']>=2010)&(df['PY']<=2012)]
# add fund label col
df1012 = df1012.AF.str.split(';', expand=True).\
                   join(df1012.fund_label).join(df1012.PY).join(df1012.IM)
df1012.fillna(value=np.nan, inplace=True)

def df_new(df):
    temp=[]
    temp_n=[]
    for row in df.iterrows():
        index, data = row
        temp.append(data.tolist())
    for i in range(len(temp)):
        for j in range(len(temp[i])-3):
            temp_n.append([temp[i][j],temp[i][-3],temp[i][-2],temp[i][-1]])
    df_n = pd.DataFrame(temp_n, columns=['author','fund','PY','IM']).dropna()
    return df_n

# total author list
dfa = df_new(df1012)
```

```python
dfa.head()
```

```python
dfa['author'] = dfa['author'].str.lstrip()
```

```python
dfa.to_csv('../data/1012authors.csv', index = False)
```

```python
#dfa=pd.read_csv('../data/Copy of 1012authors.csv')
```

```python
dfa.head()
```

```python
#the number of funding
dfa_fund = pd.pivot_table(dfa, values='fund', index=['author', 'PY'], aggfunc=np.sum)
dfa_fund_unstacked = dfa_fund.unstack()
dfa_fund_unstacked#count funding publishment
```

```python
# the number publication
dfa_pb = pd.pivot_table(dfa, values='fund', index=['author','PY'], aggfunc='count')
dfa_pb_unstacked = dfa_pb.unstack()

# the sum of impact factor
dfa_im = pd.pivot_table(dfa, index=['author', 'PY'], values='IM', aggfunc='sum')
dfa_im_unstacked = dfa_im.unstack()

# concat
dfap = pd.concat([dfa_pb_unstacked, dfa_fund_unstacked],\
                axis=1, join='outer', join_axes=[dfa_fund_unstacked.index])
```

```python
dfaf = pd.concat([dfap, dfa_im_unstacked],\
                 axis=1, join='outer', join_axes=[dfap.index])

dfa_im_unstacked.head()
```

```python
dfaf.head()#publication,funding,im
```

```python
# first author df
df1 = df1012[['PY','fund_label','IM']]
df1['first_au'] = df1012.iloc[:,0]
df1 = df1.reset_index(drop=True)
df1
```

```python
# funding
df1_fund = pd.pivot_table(df1, values='fund_label', index=['first_au', 'PY'],aggfunc=np.sum)
df1_fund_unstacked = df1_fund.unstack()

# publication
df1_pb = pd.pivot_table(df1, index=['first_au', 'PY'], values='fund_label', aggfunc='count')
df1_pb_unstacked = df1_pb.unstack()

dfap1 = pd.concat([df1_pb_unstacked, df1_fund_unstacked],\
                 axis=1, join_axes=[df1_fund_unstacked.index])

# IM
df1_im = pd.pivot_table(df1, values='IM', index=['first_au', 'PY'],aggfunc=np.sum)
df1_im_unstacked = df1_im.unstack()

df1f = pd.concat([dfap1, df1_im_unstacked],\
                 axis=1, join_axes=[dfap1.index])
```

```python
df1_fund_unstacked.head()
```

```python
df1f.head()#publish,fund, impact
```

```python
df_final = pd.concat([dfaf,df1f],axis=1,join='outer',join_axes=[dfaf.index])
```

```python
#df_final = dfap.set_index('author').join(dfap1.set_index('first_au'))
dfapi = df_final.fillna(0)
```

```python
dfapi.columns = dfapi.columns.droplevel()
```

```python
dfapi.columns=['2010pb','2011pb','2012pb',
             '2010fund','2011fund','2012fund',
            '2010im','2011im','2012im',
            '2010pbf','2011pbf','2012pbf',
            '2010fundf','2011fundf','2012fundf',
            '2010imf','2011imf','2012imf']
```

```python
dfapi.reset_index(inplace=True)
```
